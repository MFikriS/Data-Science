Koding Baris 2

```
import pandas as pd
import numpy as np

#membaca dataset yang telah diunduh 
dataset = pd.read_csv("kc_house_data.csv")

# menampilkan dataset untuk 5 baris teratas
dataset.head()
```

```
#korelasi data 

korelasi = dataset.corr()
korelasi.style.background_gradient(cmap='coolwarm')
```

Koding Baris 3

```
#menghapus fitur atau kolom pada dataset yang diperkirakan tidak berpengaruh pada klasifikasi
dataset = dataset.drop(['id','date','zipcode','long','lat','sqft_lot','condition','yr_built','sqft_lot15','yr_renovated', 'view'] ,axis=1)
```

Koding Baris 4

```
#melakukan pengisian missing value pada kolom sqft_living15 yaitu dengan cara mendeteksi missing value itu sendiri dengan fillna dan mengisi data dengan rata-rata
dataset['sqft_living15'] = dataset['sqft_living15'].fillna(dataset['sqft_living15'].mean())
#setelah dilakukukanya pengisian missing value hasil yang dilakukan dengan nilai rata-rata tadi dilakukan pembulatan angka pada kolom sqft_living15 karena agar semua data sama dengan nilai angka bulat
dataset['sqft_living15'] = round(dataset['sqft_living15'])
```

Koding Baris 5

```
#melakukan pengecekan terhadap missing value pada dataset yang telah diisi oleh bilangan bulat
dataset
```

Koding Baris 6

```
#melakukan inisiasi a untuk menyimpan dataset "price"
a = dataset.price
#membuat suatu wadah variabel mean untuk menghitung rata rata pada dataset "price"
mean = np.mean(a)
#membuat wadah standar deviasi untuk distribusi data agar terbagi dengan rata
std = np.std(a)
#membuat wadah min untuk nilai minimal pada dataset "price"
min = np.min(a)
#membuat wadah max untuk nilai maksimal pada dataset "price"
max = np.max(a)
print(mean)
print(std)
print(min)
print(max)
```

Koding Baris 7

```
import numpy as np
import scipy.stats as stats
import pylab as pl
#inisiasi h untuk mengurutkan dataset "price"
h = sorted(a)
#inisiasi fit untuk menggeser distribusi
fit = stats.norm.pdf(h, np.mean(h), np.std(h)) 
#plot untuk gambar sesuai dengan data yang telah diurutkan dan menempatkan histogram yang sesuai
pl.plot(h,fit,'-o')
#menggambar histogram untuk data yang sudah diurutkan 
pl.hist(h,normed=True)  
#menampilkan histogram 
pl.show()                   
```

Koding Baris 10

```
#melakukan pelabelan dengan nilai "price" kurang dari sama dengan 445000 diberi label 0, sedangkan jika lebih dari 445000 akan diberi label 1
def conditions(a):
    if (a['price'] <= 445000):
        return 0
    else:
        return 1
```

```
#menggabungkan dataset
dataset['label'] = dataset.apply(conditions, axis=1)
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.4)
```

Koding Baris 19

```
from sklearn.naive_bayes import GaussianNB

#inisiasi gnb yang berisi klasifikasi Gausian menggunakan library Gaussian Naive Bayes
gnb = GaussianNB()

#melatih model dengan menggunakan data training
a = gnb.fit(x_train, y_train)

#memprediksi respons untuk data testing 
y_pred = gnb.predict(x_test)
```

Koding Baris 20

```
#Import scikit-learn metrics module untuk perhitungan akurasi
from sklearn import metrics

# model akurasi berdasarkan data testing 
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
